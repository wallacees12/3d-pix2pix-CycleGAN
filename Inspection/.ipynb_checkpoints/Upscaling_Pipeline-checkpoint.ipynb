# 3D Medical Image Upscaling Pipeline (Organized Structure)

Upscale synthetic CT images generated by 3D Pix2Pix using various interpolation methods.

**New Organized Structure:**
```
results/
â””â”€â”€ model_name/
    â”œâ”€â”€ test_epoch_npz/           # NPZ inference results
    â”œâ”€â”€ for_upscaling/            # Exported synthetic CT files
    â”œâ”€â”€ upscaled_results/         # Upscaled outputs
    â””â”€â”€ web/                      # Visualization results
```

This pipeline automatically detects the model structure and processes files in the organized directory layout.

````python
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
from pathlib import Path
import time
from scipy.ndimage import zoom
from skimage.transform import resize
import SimpleITK as sitk
````

## ğŸ¯ Select Model and Load Target Resolution

````python
# Auto-detect available models in results directory
results_base = "/scratch/results/"
available_models = []

if os.path.exists(results_base):
    for item in os.listdir(results_base):
        model_path = os.path.join(results_base, item)
        if os.path.isdir(model_path):
            # Check if it has the expected structure
            for_upscaling_path = os.path.join(model_path, "for_upscaling")
            if os.path.exists(for_upscaling_path):
                available_models.append(item)

print(f"ğŸ” Available Models:")
if available_models:
    for i, model in enumerate(available_models):
        print(f"   {i}: {model}")
    
    # Select model (change index to select different model)
    model_index = 1  # Change this to select different model
    selected_model = available_models[model_index]
    
    print(f"\nğŸ¯ Selected Model: {selected_model}")
    
    # Set up paths
    model_dir = os.path.join(results_base, selected_model)
    for_upscaling_dir = os.path.join(model_dir, "for_upscaling")
    upscaled_results_dir = os.path.join(model_dir, "upscaled_results")
    
    print(f"ğŸ“ Model Directory: {model_dir}")
    print(f"ğŸ“ For Upscaling: {for_upscaling_dir}")
    print(f"ğŸ“ Upscaled Results: {upscaled_results_dir}")
    
    # Create upscaled results directory
    os.makedirs(upscaled_results_dir, exist_ok=True)
    
else:
    print("âŒ No models found with organized structure!")
    print("   Run test_npz.py first to generate organized results")
    selected_model = None

# Load target resolution from original MR/CT files
mr_path = "/Users/samwallace/Documents/sCT/Comparison Pix2Pix/3D/datasets/overfit/mr.mha"
ct_path = "/Users/samwallace/Documents/sCT/Comparison Pix2Pix/3D/datasets/overfit/ct.mha"

if os.path.exists(mr_path):
    mr_img = sitk.ReadImage(mr_path)
    target_shape = sitk.GetArrayFromImage(mr_img).shape
    print(f"\nğŸ“ Target MR shape: {target_shape}")
    
    # Load MR data for comparison if needed
    original_mr_data = sitk.GetArrayFromImage(mr_img)
    print(f"ğŸ“Š Original MR loaded: {original_mr_data.shape}")
else:
    print(f"\nâš ï¸ Original MR file not found: {mr_path}")
    print("   Using fallback target shape")
    target_shape = (161, 512, 512)  # Common high-resolution target
    original_mr_data = None
````

````python
# Load NPZ file from organized structure
if selected_model is None:
    print("âŒ Cannot proceed without a selected model")
    raise SystemExit("No model selected")

# List available NPZ files
npz_files = [f for f in os.listdir(for_upscaling_dir) if f.endswith('.npz')]
print(f"\nğŸ“„ Available NPZ files in {selected_model}:")
for i, npz_file in enumerate(npz_files):
    print(f"   {i}: {npz_file}")

if not npz_files:
    print("âŒ No NPZ files found!")
    raise SystemExit("No NPZ files found")

# Select NPZ file (change index to select different file)
npz_index = 0  # Change this to select different NPZ file
selected_npz = npz_files[npz_index]
npz_file_path = os.path.join(for_upscaling_dir, selected_npz)

print(f"\nğŸ¯ Selected NPZ: {selected_npz}")
print(f"ğŸ“„ Full path: {npz_file_path}")

# Load the NPZ data
try:
    data = np.load(npz_file_path)
    print(f"\nâœ… NPZ loaded successfully!")
    print(f"ğŸ“Š Available keys: {list(data.keys())}")
    
    # Display information about each array
    for key in data.keys():
        array = data[key]
        print(f"   {key}: shape={array.shape}, dtype={array.dtype}")
        print(f"         min={array.min():.3f}, max={array.max():.3f}, mean={array.mean():.3f}")
    
    # Extract synthetic CT data
    if 'synthetic_ct' in data:
        synthetic_ct = data['synthetic_ct']
        print(f"\nğŸ§  Synthetic CT loaded: {synthetic_ct.shape}")
    else:
        print("âŒ No 'synthetic_ct' key found in NPZ file")
        print("   Available keys:", list(data.keys()))
        # Try to use the first available array
        if len(data.keys()) > 0:
            key = list(data.keys())[0]
            synthetic_ct = data[key]
            print(f"   Using '{key}' instead: {synthetic_ct.shape}")
        else:
            raise SystemExit("No data found in NPZ file")
    
except Exception as e:
    print(f"âŒ Error loading NPZ file: {e}")
    raise SystemExit(f"Failed to load NPZ: {e}")
````

## âš™ï¸ Upscaling Configuration

````python
# Define upscaling methods to run
# Set which methods to run (True = run, False = skip)
METHODS_TO_RUN = {
    'cubic': True,
    'linear': True,
    'nearest': False,  # Usually lower quality
    'lanczos': True,   # Often gives good results
    'area': False      # Better for downscaling
}

print("ğŸ”§ Upscaling Methods Configuration:")
for method, enabled in METHODS_TO_RUN.items():
    status = "âœ… ENABLED" if enabled else "âŒ SKIPPED"
    print(f"   {method:<10}: {status}")

print(f"\nğŸ“ Current shape: {synthetic_ct.shape}")
print(f"ğŸ“ Target shape:  {target_shape}")

# Calculate scale factors
current_shape = synthetic_ct.shape
scale_factors = tuple(targ / curr for curr, targ in zip(current_shape, target_shape))
print(f"ğŸ“Š Scale factors: {scale_factors}")

# Initialize results dictionary
upscaled_results = {}

# Base filename for saving (extracted from NPZ filename)
base_filename = selected_npz.replace('.npz', '')
print(f"\nğŸ’¾ Base filename: {base_filename}")
print(f"ğŸ“ Output directory: {upscaled_results_dir}")

saved_files = []
````

## ğŸ”§ Upscaling Methods

````python
# Perform upscaling for each enabled method
for method, enabled in METHODS_TO_RUN.items():
    if not enabled:
        print(f"â­ï¸  Skipping {method}")
        continue
    
    print(f"\nğŸ”„ Processing with {method} interpolation...")
    
    try:
        # Convert numpy array to SimpleITK image
        sitk_image = sitk.GetImageFromArray(synthetic_ct)
        
        # Set original spacing (assuming isotropic for synthetic data)
        original_spacing = (1.0, 1.0, 1.0)  # Can be adjusted if known
        sitk_image.SetSpacing(original_spacing)
        
        # Calculate new spacing based on scale factors
        new_spacing = tuple(orig_sp / scale for orig_sp, scale in zip(original_spacing, scale_factors))
        
        # Perform resampling
        resample = sitk.ResampleImageFilter()
        resample.SetSize([int(target_shape[2]), int(target_shape[1]), int(target_shape[0])])  # SimpleITK uses x,y,z order
        resample.SetOutputSpacing(new_spacing)
        resample.SetOutputOrigin(sitk_image.GetOrigin())
        resample.SetOutputDirection(sitk_image.GetDirection())
        
        # Set interpolation method
        if method == 'cubic':
            resample.SetInterpolator(sitk.sitkBSpline)
        elif method == 'linear':
            resample.SetInterpolator(sitk.sitkLinear)
        elif method == 'nearest':
            resample.SetInterpolator(sitk.sitkNearestNeighbor)
        elif method == 'lanczos':
            resample.SetInterpolator(sitk.sitkLanczosWindowedSinc)
        elif method == 'area':
            # Area interpolation not directly available in SimpleITK, use linear as fallback
            resample.SetInterpolator(sitk.sitkLinear)
            print(f"   âš ï¸  Area interpolation not available, using linear")
        
        # Apply resampling
        upscaled_image = resample.Execute(sitk_image)
        
        # Convert back to numpy array
        result = sitk.GetArrayFromImage(upscaled_image)
        
        # Clip negative values to -1024 HU (air threshold in CT imaging)
        original_min = result.min()
        result = np.clip(result, -1024, result.max())
        clipped_min = result.min()
        
        # Store result
        upscaled_results[method] = result
        
        print(f"   âœ… {method}: {synthetic_ct.shape} â†’ {result.shape}")
        print(f"   ğŸ“Š Value range: {result.min():.3f} to {result.max():.3f}")
        if original_min < -1024:
            print(f"   ğŸ”§ Clipped {(original_min < -1024).sum() if hasattr(original_min, 'sum') else 'some'} values from {original_min:.1f} to -1024 HU")
        
        # Save as .nii file in organized directory
        output_filename = f"{base_filename}_{method}_upscaled.nii"
        output_path = os.path.join(upscaled_results_dir, output_filename)
        
        # Convert clipped result back to SimpleITK image for saving
        clipped_sitk_image = sitk.GetImageFromArray(result)
        clipped_sitk_image.SetSpacing(new_spacing)
        clipped_sitk_image.SetOrigin(sitk_image.GetOrigin())
        clipped_sitk_image.SetDirection(sitk_image.GetDirection())
        
        sitk.WriteImage(clipped_sitk_image, output_path)
        saved_files.append(output_path)
        print(f"   ğŸ’¾ Saved: {output_filename}")
        
    except Exception as e:
        print(f"   âŒ Error with {method}: {e}")
        continue

print(f"\nğŸ‰ Upscaling complete!")
print(f"ğŸ“ Results saved in: {upscaled_results_dir}")
print(f"ğŸ’¾ Files created:")
for file_path in saved_files:
    filename = os.path.basename(file_path)
    print(f"   ğŸ“„ {filename}")

# Display summary statistics
print(f"\nğŸ“Š Summary:")
print(f"   Original shape: {synthetic_ct.shape}")
print(f"   Target shape:   {target_shape}")
print(f"   Methods run:    {len(upscaled_results)}")
for method, result in upscaled_results.items():
    print(f"   {method:<10}: shape={result.shape}, range=[{result.min():.2f}, {result.max():.2f}]")
print(f"\nğŸ¥ Note: All results clipped to minimum -1024 HU (air threshold)")
````

## ğŸš€ Run Upscaling

````python
# Load original high-resolution data for comparison
mr_img = sitk.ReadImage(mr_path)
original_mr_for_comparison = sitk.GetArrayFromImage(mr_img)
ct_img = sitk.ReadImage(ct_path)
original_ct_for_comparison = sitk.GetArrayFromImage(ct_img)
````

## ğŸ“Š Compare Upscaling Results

````python
# Create comprehensive visualization
if upscaled_results:
    # Define slice index for consistent comparison (middle slice)
    slice_idx = synthetic_ct.shape[0] // 2
    print(f"ğŸ” Using slice {slice_idx} for comparison (middle of {synthetic_ct.shape[0]} slices)")
    
    num_methods = len(upscaled_results)
    # Calculate grid size: compare original low-res, upscaled results, and ground truth
    num_comparisons = 1 + num_methods + (1 if original_ct_for_comparison is not None else 0)
    
    fig, axes = plt.subplots(1, num_comparisons, figsize=(4*num_comparisons, 6))
    if num_comparisons == 1:
        axes = [axes]
    
    # Plot original low-resolution synthetic CT
    ax_idx = 0
    im0 = axes[ax_idx].imshow(synthetic_ct[slice_idx], cmap='gray')
    axes[ax_idx].set_title(f'Original Synthetic CT\n{synthetic_ct.shape}')
    axes[ax_idx].axis('off')
    plt.colorbar(im0, ax=axes[ax_idx], fraction=0.046, pad=0.04)
    ax_idx += 1
    
    # Plot upscaled results
    for method, result in upscaled_results.items():
        # Use corresponding slice index for upscaled results
        upscaled_slice_idx = result.shape[0] // 2
        im = axes[ax_idx].imshow(result[upscaled_slice_idx], cmap='gray')
        axes[ax_idx].set_title(f'{method.capitalize()} Upscaled\n{result.shape}')
        axes[ax_idx].axis('off')
        plt.colorbar(im, ax=axes[ax_idx], fraction=0.046, pad=0.04)
        ax_idx += 1
    
    # Plot ground truth if available
    if original_ct_for_comparison is not None:
        gt_slice_idx = original_ct_for_comparison.shape[0] // 2
        im_gt = axes[ax_idx].imshow(original_ct_for_comparison[gt_slice_idx], cmap='gray')
        axes[ax_idx].set_title(f'Ground Truth CT\n{original_ct_for_comparison.shape}')
        axes[ax_idx].axis('off')
        plt.colorbar(im_gt, ax=axes[ax_idx], fraction=0.046, pad=0.04)
    
    plt.tight_layout()
    plt.suptitle(f'Upscaling Results Comparison - {selected_model}\nSlice {slice_idx} (original) / middle slices (others)', y=1.02, fontsize=14)
    
    # Save comparison plot
    plot_filename = f"{base_filename}_upscaling_comparison.png"
    plot_path = os.path.join(upscaled_results_dir, plot_filename)
    plt.savefig(plot_path, dpi=150, bbox_inches='tight')
    print(f"ğŸ’¾ Comparison plot saved: {plot_filename}")
    
    plt.show()
    
    # Print detailed comparison
    print(f"\nğŸ“Š Detailed Comparison:")
    print(f"   Original shape: {synthetic_ct.shape}")
    print(f"   Original range: [{synthetic_ct.min():.3f}, {synthetic_ct.max():.3f}]")
    
    for method, result in upscaled_results.items():
        print(f"   {method:<10}: {result.shape}, range=[{result.min():.3f}, {result.max():.3f}]")
    
    if original_ct_for_comparison is not None:
        print(f"   Ground truth:  {original_ct_for_comparison.shape}, range=[{original_ct_for_comparison.min():.3f}, {original_ct_for_comparison.max():.3f}]")
        
        # Calculate some basic metrics if shapes match
        for method, result in upscaled_results.items():
            if result.shape == original_ct_for_comparison.shape:
                
                mse = np.mean((result - original_ct_for_comparison) ** 2)
                psnr = 20 * np.log10(np.max(original_ct_for_comparison) / np.sqrt(mse)) if mse > 0 else float('inf')
                print(f"   {method} vs GT - MSE: {mse:.3f}, PSNR: {psnr:.2f} dB")
    
else:
    print("âŒ No upscaled results to visualize")
````

## ğŸ’¾ Save Upscaled Results

````python
# Save all upscaled results as .nii files in upscaled_results_dir
for method, result in upscaled_results.items():
    nii_image = sitk.GetImageFromArray(result)
    output_filename = f"{base_filename}_{method}_upscaled.nii"
    output_path = os.path.join(upscaled_results_dir, output_filename)
    sitk.WriteImage(nii_image, output_path)# Save all upscaled results as .nii files in upscaled_results_dir
for method, result in upscaled_results.items():
    nii_image = sitk.GetImageFromArray(result)
    output_filename = f"{base_filename}_{method}_upscaled.nii"
    output_path = os.path.join(upscaled_results_dir, output_filename)
    sitk.WriteImage(nii_image, output_path)
````
